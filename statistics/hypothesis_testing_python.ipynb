# Statistics
  
## Hypothesis Testing
  
### Claim procedure:
1. Interpret the claim
- State the claim symbolically
- State the opposite of the original claim as well
The original claim could be $H_0$ or $H_1$. Where the equality is determines which is $H_0$
ex. The proportion of male CEOs is greater than .5  

claim: p > .5 -> $H_1$: P > .5   
opp:   p $\leq$ .5  -> $H_0$: P = .5  

2. Find the test statistic
Proportions - Always uses Z
$$ Z = \frac{\hat{p}-p}{\sqrt{\frac{p*q}{n}}}
$$
Mean - If you know the population stdev
$$ Z = \frac{\bar{X}-\mu}{\frac{\sigma}{\sqrt{n}}}
$$
Mean - If you only have sample stdev
$$ T = \frac{\bar{X}-\mu}{\frac{s}{\sqrt{n}}}
$$

Example:  
A sample of 706 companies fond 61% of CEOs were male. The claim is that most CEOs are male.
